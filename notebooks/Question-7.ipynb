#### Average number of aces per match

```python
import pandas as pd
import numpy as np
from glob import glob
import os
```

```python
days_to_query = os.listdir("../data/raw/tennis_data")
statistics_load_files = []

for day in days_to_query:
    statistics_load_files.extend(
        glob(
            f"../data/raw/tennis_data/{day}/data/raw/raw_statistics_parquet/*.parquet"
        )
    )

stat_df = pd.concat([pd.read_parquet(file) for file in statistics_load_files], ignore_index=True)
display(stat_df)
```

```python
aces_df = stat_df[stat_df["statistic_name"] == 'aces'].copy()
count_aces = aces_df["statistic_name"].count()
print(f"Total amount of aces: {count_aces}")

total_match = stat_df['match_id'].drop_duplicates().count()
print(f"Total amont of match: {total_match}")

aces_df['total_aces'] = aces_df['home_value'] + aces_df['away_value']
aces_per_match = aces_df.groupby('match_id')['total_aces'].sum()
mean_aces = aces_per_match.mean()

print("Average number of aces per match:", mean_aces)
```
