#### The highest winning percentage against top 10 ranked opponents

```python
import pandas as pd
import numpy as np
from glob import glob
import os
```

```python
days_to_query = os.listdir("../data/raw/tennis_data")
event_load_files = []

for day in days_to_query:
    event_load_files.extend(
        glob(
            f"../data/raw/tennis_data/{day}/data/raw/raw_match_parquet/event_*.parquet", recursive=True
        )
    )
eventinfo_df = pd.concat([pd.read_parquet(file) for file in event_load_files], ignore_index=True)
display(eventinfo_df)
display(eventinfo_df.info())
```

```python
days_to_query_homeplayer = os.listdir("../data/raw/tennis_data")
h_palyer_load_files = []

for day in days_to_query_homeplayer:
    h_palyer_load_files.extend(
        glob(
            f"../data/raw/tennis_data/{day}/data/raw/raw_match_parquet/home_team_*.parquet"
        )
    )
homeplayer_df = pd.concat([pd.read_parquet(file) for file in h_palyer_load_files], ignore_index=True)
display(homeplayer_df)
```

```python
days_to_query_awayplayer = os.listdir("../data/raw/tennis_data")
a_palyer_load_files = []

for day in days_to_query_awayplayer:
    a_palyer_load_files.extend(
        glob(
            f"../data/raw/tennis_data/{day}/data/raw/raw_match_parquet/home_team_*.parquet"
        )
    )
awayplayer_df = pd.concat([pd.read_parquet(file) for file in a_palyer_load_files], ignore_index=True)
display(awayplayer_df)
```

```python
totalplayer_df = pd.concat([homeplayer_df,awayplayer_df])
mergedinfo_df = eventinfo_df.merge(totalplayer_df[['match_id','full_name','player_id','current_rank']], on= 'match_id')

winnerplayer_df = mergedinfo_df[mergedinfo_df['winner_code'] == 1]
countwinner_player = winnerplayer_df.groupby(['full_name','player_id','current_rank']).agg(countwinner_player=('winner_code','size')).reset_index()
display(countwinner_player)
```

```python
topwinner_player = countwinner_player[countwinner_player['current_rank'] <= 10]
specific_palyer = topwinner_player.drop_duplicates(subset='player_id')
top_palyer = specific_palyer.drop_duplicates(subset='current_rank')
sort_top_player = top_palyer.sort_values('countwinner_player', ascending=False).reset_index()
display(sort_top_player)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create color column (first row = red, rest = gray)
sort_top_player['color'] = ['#AD1357' if i == 0 else '#96CBFC' for i in range(len(sort_top_player))]

# Plot using Seaborn (default color, we'll set actual bar colors manually)
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    data=sort_top_player,
    x='countwinner_player',  
    y='full_name',
    color='gray'  # Temporary uniform color
)

# Set each bar's color manually
for bar, color in zip(ax.patches, sort_top_player['color']):
    bar.set_facecolor(color)

sns.despine(top=True, right=True)
# Improve appearance
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Ranked Players by Win Count')
plt.xlabel('Player Name')
plt.ylabel('Win Count')
plt.tight_layout()
plt.show()
```
