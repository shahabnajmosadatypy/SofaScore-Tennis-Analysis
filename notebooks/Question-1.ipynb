### Amount of Tennis players in tournament

#### Import libraries

```python
import pandas as pd
import numpy as np
from glob import glob
import os
```

#### Read .parquet files
###### Count amount of players based on request in docs/Tennis project.pdf 

```python
days_to_query_home = os.listdir("../data/raw/tennis_data")
load_files_home = []

for days in days_to_query_home:
    load_files_home.extend(
        glob(
            f"../data/raw/tennis_data/raw_match_parquet/home_team_1*.parquet"
        )
    )

home_team_df = pd.concat(
    [pd.read_parquet(file,columns=["full_name","player_id"])
    for file in load_files_home],
    ignore_index=True)

display(home_team_df)
```

```python
days_to_query_away = os.listdir("../data/raw/tennis_data")
load_files_away = []

for days in days_to_query_away:
    load_files_away.extend(
        glob(
            f"../data/raw/tennis_data/raw_match_parquet/away_team_1*.parquet"
        )
    )
away_team_df = pd.concat(
    [pd.read_parquet(
        file,
        columns= ["full_name","player_id"]
    ) for file in load_files_away],
    ignore_index=True
)

display(away_team_df)
```

##### Combine both data frame to calculate amount of players

```python
# combine both data frame
total_team_df = pd.concat([home_team_df,away_team_df])

# drop duplicates player id and set a unique data frame then reset index
total_unique = total_team_df[["full_name","player_id"]].drop_duplicates(subset='player_id').reset_index(drop=True)

# count amount of players
counter = total_unique["player_id"].count()
print(f"The amount of player: {counter}")
```
