## Amount of tennis player in dataset
#### Import libraries

```python
import pandas as pd
from glob import glob
import os
```

```python
match_parquet_path = "D:/Daneshkar/Project-1(Tennis)/data/input/tennis_data/**/data/raw/raw_match_parquet/*.parquet"


match_parquet_files = glob(match_parquet_path, recursive=True)
limited_files = sorted(set(os.path.dirname(f) for f in match_parquet_files))
selected_days = limited_files[:]
parquet_file_selected = [file for file in match_parquet_files if os.path.dirname(file) in selected_days]
combined_df = pd.concat([pd.read_parquet(f) for f in parquet_file_selected], ignore_index=True)
display(combined_df)
```

```python
combined_df.info()
```

### Answer Question 1

```python
# Filter rows where player_id is not null
filtered_df = combined_df[combined_df['player_id'].notna()]

# Print name and player_id columns
unique_id = filtered_df[['name', 'player_id']].drop_duplicates()
print(unique_id)

# print player count
print(f"\n{unique_id['player_id'].count()} players attended to the tournament")
```
